```python
#Importing libraries
import pandas as pd
```

```python
#Loading data
data = pd.read_csv('C:/Users/bhavi/OneDrive/Desktop/Data/Time_Series_Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
#Checking data
data.head()
```

```python
#Creating plot on close features
data['Close'].plot(figsize = (12,5));
```

# Rolling (Creating moving average)

```python
#Extracting 7 days moving average from the data
data.rolling(window=7).mean()
```

```python
#Creating plot on close features & its 7 days moving average
data['Close'].plot(figsize = (12,5));
data['Close'].rolling(window=7).mean().plot();
```

```python
#Creating plot on close features & its 30 days moving average
data['Close'].plot(figsize = (12,5));
data['Close'].rolling(window=30).mean().plot();
```

```python
#Creating plot on close features & its 60 days moving average
data['Close'].plot(figsize = (12,5));
data['Close'].rolling(window=60).mean().plot();
```

```python
#Adding 30 days moving average to the data
data['Close_30 days moving average'] = data['Close'].rolling(window=30).mean()
```

```python
#Checking data
data.head(50)
```

```python
#Ploting Close and Close_30 days moving average
data[['Close', 'Close_30 days moving average']].plot(figsize = (16,9));
```

# Expanding(Creating cummulative adding data)

```python
#Expanding the data (Cummulative adding the data)
data['Close'].expanding().mean()
```

```python
#Plotting expanding the data (Cummulative adding the data)
data['Close'].expanding().mean().plot(figsize = (12,5));
```
