# Plotting Close Feature & Volume Feature

```python
#Plotting Close feature
plt.subplot(2,1,1)
data['Close'].plot(figsize = (16,9), title = 'Close');

#Plotting Volumn feature
plt.subplot(2,1,2)
data['Volume'].plot(figsize = (16,9), title = 'Volume');
plt.subplots_adjust(hspace=0.5)
```

# Plotting Close Feature with additional plotting parameter

```python
#Creating plot on Close feature
ax = data['Close'].plot(figsize = (12,6), title = 'Close');
ax.autoscale(axis = 'both', tight = True)
ax.set(xlabel = 'Year', ylabel = 'Close');
```

# Plotting Close Feature with selecting periods

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
data['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,4));
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
data['Close'].plot(figsize=(12,4),xlim=['2017-01-01','2017-12-31']);
```

# Plotting Close Feature with Changing size, y axis limit, style of line & color of plot

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
data['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,4), #Changing size of the plot
                                              ylim=[0,70], #Changing the y axis limit
                                              ls = '--', #Changing style of the line
                                              c = 'green'); #Changing color of the line
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
data['Close'].plot(figsize=(12,4),
                   xlim=['2017-01-01','2017-03-01'], 
                   ylim = [50,60]);
#It has wide gaps on x axis
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
ax = data['Close'].plot(figsize=(12,4), 
                        xlim=['2017-01-01','2017-03-01'], 
                        ylim = [50,60]);
ax.set(xlabel ='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d'))
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
ax = data['Close'].plot(figsize=(12,4), xlim=['2017-01-01','2017-03-01'], ylim = [50,60]);
ax.set(xlabel ='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%b'))
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
ax = data['Close'].plot(figsize=(12,4), xlim=['2017-01-01','2017-03-01'], ylim = [50,60]);
ax.set(xlabel ='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))
```

```python
#Plotting Close features between 2017-01-01 to 2017-12-31
ax = data['Close'].plot(figsize=(12,4), xlim=['2017-01-01','2017-03-01'], ylim = [50,60]);
ax.set(xlabel ='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
