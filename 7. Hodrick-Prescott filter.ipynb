```python
#Importing libraries for data manipulation
import numpy as np
import pandas as pd

#Importing libraries for data plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#Importing libraries for Hodrick-Prescott filter
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
#Loading data
data = pd.read_csv('C:/Users/bhavi/OneDrive/Desktop/Data/Time_Series_Data/macrodata.csv', 
                   index_col=0, 
                   parse_dates=True)
```

```python
#Checking the data
data.head()
```

```python
#Plotting realgdp
data['realgdp'].plot(figsize = (12,5));
```

```python
#Creating trend and cycle series from data
gdp_cycle, gdp_trend = hpfilter(data['realgdp'], lamb=1600)
```

```python
#Checking the type of result
type(gdp_cycle)
```

```python
#Creating plot on gdp cycle and trend
plt.subplot(1,2,1)
gdp_cycle.plot(figsize = (12,5));
plt.title('GDP Cycle')

#Creating plot on gdp cycle and trend
plt.subplot(1,2,2)
gdp_trend.plot();
plt.title('GDP Trend');
```

```python
#Comparing real GDP and Trend 
data['realgdp'].plot(figsize = (12,5));
gdp_trend.plot();
plt.title('Real GDP vs GDP Trend');
```

```python
#Creating plot on realgdp and cycle
data['realgdp'].plot(figsize = (12,5));
gdp_cycle.plot();
plt.title('Real GDP vs GDP Cycle');
```

```python
#Creating plot on realgdp and trend from 2005 onwards
data['realgdp']['2005-01-01':].plot(figsize = (12,5));
gdp_trend['2005-01-01':].plot();
plt.title('Real GDP vs GDP Trend from 2015 onwards');
```
