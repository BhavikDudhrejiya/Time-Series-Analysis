```python
#Importing libraries
import pandas as pd
import numpy as np
%matplotlib inline
from statsmodels.tools.eval_measures import mse, rmse, meanabs
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
```

```python
#Creating data
np.random.seed(42)
data = pd.DataFrame(np.random.randint(20,30,(50,2)), columns=['Test', 'Predictions'])
```

```python
#Checking data
data.head()
```

```python
#Plotting data
data.plot(figsize = (12,8));
```

```python
#Computing mean squared error
mse(data['Test'], data['Predictions'])
```

```python
#Computing root mean squared error
rmse(data['Test'], data['Predictions'])
```

```python
#Computing absolute mean error
meanabs(data['Test'], data['Predictions'])
```
